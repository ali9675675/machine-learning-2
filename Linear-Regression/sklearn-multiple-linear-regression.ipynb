# Multiple linear regression

## Import the relevant libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
from sklearn.linear_model import LinearRegression
```

## Load the data

```python
data = pd.read_csv('007 1.02.Multiple-linear-regression[UdemyIran.Com].csv')
data.head(5)
```

```python
data.describe()
```

## Create the multiple linear regression

### Declare the dependent and independent variables

```python
x = data[['SAT', 'Rand 1,2,3']]
y = data['GPA']
```

### Regression itself

```python
reg = LinearRegression()
reg.fit(x, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

## Calculating R-squared

```python
reg.score(x,y)
```

### Formula for Adjusted R^2

$R^2_{adj.} = 1 - (1-R^2)*\frac{n-1}{n-p-1}$

```python
x.shape
```

```python
r2 = reg.score(x,y)
n = x.shape[0]
p=x.shape[1]
adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
adjusted_r2
```

## Feature selection

```python
from sklearn.feature_selection import f_regression
```

```python
f_regression(x,y)
```

```python
p_values = f_regression(x,y)[1]
p_values
```

```python
p_values.round(3)
```

## Creating a summary table

```python
reg_summary = pd.DataFrame(data=['SAT', 'Rand 1,2,3'], columns=['Features'])
reg_summary
```

```python
reg_summary = pd.DataFrame(data=x.columns.values, columns=['Features'])
reg_summary
```

```python
reg_summary['Coefficients'] = reg.coef_
reg_summary['p-values'] = p_values.round(3)
```

```python
reg_summary
```

## Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

## Regression with scaled features

```python
reg.fit(x_scaled,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

## Creating a summary table

```python
reg_summary = pd.DataFrame([['Bias'], ['SAT'], ['Rand 1,2,3']], columns=['Features'])
reg_summary['Weights'] = reg.intercept_, reg.coef_[0], reg.coef_[1]
```

```python
reg_summary
```

## making predictions with the standardized coefficients (weights)

```python
new_data = pd.DataFrame(data=[[1700,2],[1800,1]], columns=['SAT', 'Rand 1,2,3'])
new_data
```

```python
reg.predict(new_data)
```

```python
new_data_scaled = scaler.transform(new_data)
new_data_scaled
```

```python
reg.predict(new_data_scaled)
```

## What if we removed the 'Random 1,2,3' variable?

```python
reg_simple = LinearRegression()
x_simple_matrix = x_scaled[:,0].reshape(-1,1)
reg_simple.fit(x_simple_matrix,y)
```

```python
reg_simple.predict(new_data_scaled[:,0].reshape(-1,1))
```
